```python
selected_columns = train_data[required_cols]
train_data_clean_X = selected_columns.copy()
#train_data_clean_without_ind
train_data_clean_X.drop(['default_ind','application_key'], axis = 1,inplace=True)
train_data_clean_X
```

```python
#splitting train and test data
from sklearn.model_selection import train_test_split
#X = df[required_cols]
X = train_data_clean_X
y = train_data['default_ind']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
decision_tree = DecisionTreeClassifier(criterion='gini',max_depth=10)
decision_tree.fit(X_train,y_train)
predictions_train = decision_tree.predict(X_train)
train_accuracy = accuracy_score(y_train,predictions_train)
print(train_accuracy)
predictions_test = decision_tree.predict(X_test)
test_accuracy = accuracy_score(y_test,predictions_test)
print(test_accuracy)
```
