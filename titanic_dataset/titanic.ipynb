```python
# ohe
def ohe(self,string=None):
    """
    One hot encoding for the 'recent_trend' feature
    """
    if string == 'removed_nans':
        dat = self.remove_nans()
    else:
        dat = self.replace_nans()
    enc = OneHotEncoder(sparse=False)
    enc.fit(dat['recent_trend'].to_numpy().reshape(-1,1))
    dat[['falling', 'rising', 'stable']] = enc.transform(dat['recent_trend'].to_numpy().reshape(-1,1))
    #data1[['falling', 'rising', 'stable']] = enc.transform(data1['recent_trend'].to_numpy().reshape(-1,1))

    dat.drop(columns = ['recent_trend'],inplace=True)
    #data1.drop(columns = ['recent_trend'],inplace=True)
    return(dat)
```
